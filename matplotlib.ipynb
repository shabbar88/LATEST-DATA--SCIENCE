```python
over_50['target'].values
```

```python
plt.style.use("seaborn-whitegrid")
#oo methhod from scratch
fig,ax=plt.subplots(figsize=(10,6))
# plotting data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter");

# customise plot
ax.set(title="Heart Disease and Cholestral levels",xlabel="Age",ylabel="cholestral");
ax.set_xlim([45,100]);
# add legend
ax.legend(*scatter.legend_elements(),title="Target");
# add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle="--");

```

```python
plt.style.use("seaborn-whitegrid")
# subplot of chol,age,thalac
fig,(ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)

# add data to ax0
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter");

# customise plot
ax0.set(title="Heart Disease and Cholestral levels",ylabel="cholestral");
ax0.set_xlim([45,100]);
# add legend
ax0.legend(*scatter.legend_elements(),title="Target");
# add horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle="--");


# add data to ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap="summer");

# customise plot
ax1.set(title="Heart Disease and Max heart rate",xlabel="Age",ylabel="Max heart rate");
ax1.set_xlim([45,100]);
# add legend
ax1.legend(*scatter.legend_elements(),title="Target");
# add horizontal line
ax1.axhline(over_50['thalach'].mean(),linestyle="--");
# add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight='bold');



```
