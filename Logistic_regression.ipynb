```python
TP=len(y_pred_df[(y_pred_df['actual']==1) & (y_pred_df['predictions']==1)])
TN=len(y_pred_df[(y_pred_df['actual']==0) & (y_pred_df['predictions']==0)])
FP=len(y_pred_df[(y_pred_df['actual']==0) & (y_pred_df['predictions']==1)])
FN=len(y_pred_df[(y_pred_df['actual']==1) & (y_pred_df['predictions']==0)])
```

```python
print('True Positives', TP)
print('True Negatives', TN)
print('False Positives', FP)
print('False Negatives', FN)
```

If TRUE POSITIVES are 289, it means in the case of 289 Passengers, the model prediction was right while it says passengers will be survived.

If TRUE NEGATIVES are 498, it means in the case of 312 passengers, the model prediction was right while it says passengers will not survived.

If FALSE POSITIVES are 90, it means in the case of 154 passengers, the model prediction was wrong while it says passengers will survive.

If FALSE NEGATIVES are 119, it means in the case of 199 passengers, the model prediction was wrong while it says passengers will not survive.

Null Hypothesis : passenger survive (1)
Alternate Hypothesis : passenger not survive (0)
Type 1 error will be when we accept alternate hypothesis when null is true, therefore FALSE NEGATIVE in this case

Type 2 error will be when we reject alternate hypothesis when null is false, it is a FALSE POSITIVE in this case

```python
Result=[]
for i in range(len(y_pred_df)):
    if (y_pred_df.loc[i,'actual']==1) & (y_pred_df.loc[i,'predictions']==1):
      Result.append('True Positive')
    elif (y_pred_df.loc[i,'actual']==0) & (y_pred_df.loc[i,'predictions']==0):
      Result.append('True Negative')
    elif (y_pred_df.loc[i,'actual']==1) & (y_pred_df.loc[i,'predictions']==0):
      Result.append('False Negative')
    elif (y_pred_df.loc[i,'actual']==0) & (y_pred_df.loc[i,'predictions']==1):
      Result.append('False Positive')
    else:
      Result.append('Invalid')

y_pred_df['Result']=pd.Series(Result)
y_pred_df[0:30]

#y_pred_df.drop('index',axis=1,inplace=True)
y_pred_df
```

Sensitivity = TP/(TP+FN)

Specificity = TN/(FP+TN)

Prevalence = (TP+FN)/(TP+FP+FN+TN)

PPV = (sensitivity * prevalence)/((sensitivityprevalence) + ((1-specificity)(1-prevalence)))

NPV = (specificity * (1-prevalence))/(((1-sensitivity)prevalence) + ((specificity)(1-prevalence)))

Detection Rate = TP/(TP+FP+FN+TN)

Detection Prevalence = (TP+FP)/(TP+FP+FN+TN)
Balanced Accuracy = (sensitivity+specificity)/2
Precision (Positive Predicted Value) = TP/(TP+FP)

Recall = TP/(TP+FN)

Negative Predicted Value = TN/(FN+TN)

Miss Rate(False Negative Rate) = FN/(TP+FN)

Fall Out (False Positive Rate) = FP/(FP+TN)

False Discovery Rate = FP/(TP+FP)

False Omission Rate = FN/(FN+TN)

Sensitivity = (Actual 1's & Predicted 1's)/ Total Actual 1's

Specificity = (Actual 0's & Predicted 0's)/ Total Actual 0's

Precision = (Actual 1's & Predicted 1's)/ Total Predicted 1's

Sensitivity is 0.71, which means out of total number of passengers who survived  (Label 1), in 71% cases the model could predict them correctly.

Specificity is 0.84, which means out of total number of clients who actually adid not survived (Label 0), in 84% cases the model could predict them correctly.

Precision is 0.76, which means out of total number of clients where model says that they passenger survived, it was right in saying so in 76% of cases.
